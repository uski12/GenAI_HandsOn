```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
def generate_youtube_titles(niche):
    prompt = f"""
    {niche} List of viral video titles are
    """

    output = generator(
        prompt,
        max_length=120,
        num_return_sequences=1,
        temperature=0.95,
        top_p=0.95,
        do_sample=True
    )

    text = output[0]["generated_text"]

    # Clean and extract titles
    titles = []
    for line in text.split("\n"):
        line = line.strip()
        if line and line[0].isdigit():
            titles.append(line)

    return titles[:5]
```

```python
niche = "Cars"
titles = generate_youtube_titles(niche)

for t in titles:
    print(t)
```

# Observations
GPT2 is unable to "follow" instructions per-se. In tasks where it only has to generate the next token based on the previous token, without following any pre-defined rules, it does pretty well (although still not always meaningful).

After trying out various different prompts, such as:

List of viral video titles for {niche}:

1.

GPT2 simply outputs: "1.\n2.\n3.\n4.\n5."

This confirms that it is unable to follow instructions, but rather merely follows the pattern.

Gives us "better" answers (as shown in the output of the cell).
